```python
#Alejandro Morales Madrigal
#Autocorrelation

import matplotlib.pyplot as plt
import random
import math
```

```python
A = [0]
x = [0]
autocorr = []

#Generates a Path with a uniform random contribution
for i in range(1,2000):
    x.append(i)
    A.append(A[i-1]+((i+1)*.0001)+(500*random.uniform(-1,1)))
    #Para que veas que cuando no se van sumando la autocorrelacion en 0 para puros random
    #A.append((500*random.uniform(-1,1)))

plot1 = plt.figure(1,figsize=(6,3), dpi=100)    
plt.plot(x,A,'g')

#Calculating the mean
sum = 0
for j in range(len(A)):
    sum = sum + A[j]

mean = sum/len(A)

#Calculating the Standard Deviation
var = 0
for k in range(len(A)):
    var = var + ((A[k]-mean)*(A[k]-mean))

var = var/len(A)
SD = math.sqrt(var)

x1 = []
#Autocorrelation up to 1000 lags
for m in range(1000):   
    x1.append(m)
    lag = 0
    for l in range(m,len(A)):
        lag = lag + ((A[l-m]-mean)*(A[l]-mean))
    autocov = lag/len(A)
    autocorr.append(autocov/var)

plot2 = plt.figure(2,figsize=(6,4), dpi=100) 
plt.bar(x1,autocorr)
```
